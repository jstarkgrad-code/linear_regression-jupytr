This notebook demonstrates a simple linear regression analysis using Python. to model Salary based on Years of Experience.

```python
import pandas as pd
dataset= pd.read_csv("regression_data.csv")
```

this line imports the pandas library and renames it as "pd". Then creates a variable called "dataset" and assigns the title to CSV file, read with pandas, called "regression_data.csv", that is located in the directory. Finally, it reads the contents of the CSV file and saves them as a table

```python
import matplotlib.pyplot as plt
plt.scatter(dataset["YearsExperience"], dataset["Salary"], color="red")
```

This line uses the matplot library pyplot interface, renames it to "plt" then, uses it to create a scatter plot using the "dataset" variable from line one. Assigns the columns from the dataset to an axis: x = years experience and y = salary. It uses the color red to make the plot points. 

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(dataset[["YearsExperience"]], dataset[["Salary"]])
```

This section accesses the package scikit-learn and imports the a Linear Regression class. The variable "model" is created and assigned the function LinearRegression(). Finally, the linear regression model is fit to the previously defined dataset with Years of Experience as the x variable and Salary as the y variable.

```python
plt.scatter(dataset["YearsExperience"], dataset["Salary"], color="red")
plt.plot(dataset["YearsExperience"], model.predict(dataset[["YearsExperience"]]), color="blue")
plt.title("Salary vs Years Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

The next part uses the fitted linear regression model from the previous section and overlays the predicted regression line in blue on the matplotlib scatter plot of the same dataset. Note that I wanted the scatter plot's individual datapoints to still be visible so included plt.scatter again. Finally, it creates axis titles and chart title from inputs and creates the figure. 

```python
model.score(dataset[["YearsExperience"]], dataset[["Salary"]]), # R-squared
```

This line evaluates the model by outputing an R^2 value
