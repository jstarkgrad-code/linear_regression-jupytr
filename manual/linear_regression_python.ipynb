This notebook demonstrates a simple linear regression analysis using Python. to model Salary based on Years of Experience.

this line imports the pandas, matplot, and sklearn libraries and renames them for easier calling. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

This section reads a CSV file, hardcoded, and names it "dataset" for easier referal. Then calls the LinearRegression() function. Finally, the linear regression model is fit to the previously defined dataset with Years of Experience as the x variable and Salary as the y variable.

```python
dataset= pd.read_csv("regression_data.csv")
model = LinearRegression()
model.fit(dataset[["YearsExperience"]], dataset[["Salary"]])
```

The next section calculates mean squared error using sklearn.metrics import and mean_squared_error function

```python
y = dataset["Salary"]
y_predict = model.predict(dataset[["YearsExperience"]])
mse = mean_squared_error(y, y_predict)
```

The next part uses the fitted linear regression model from the previous section and overlays the predicted regression line in blue on the matplotlib scatter plot of the same dataset. Next, creates axis titles and chart title from inputs and creates the figure. Finally, creates text overlay on the figure starting at x point 2.7 and y point 40,000 that displays the slope-intercept formula, R^2 value, and mean squared error.

```python
plt.scatter(dataset["YearsExperience"], dataset["Salary"], color="red")
plt.plot(dataset["YearsExperience"], model.predict(dataset[["YearsExperience"]]), color="blue")
plt.title("Salary vs Years Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.text(
	2.7, 40000, 
	f"y = {model.coef_[0, 0]:.2f}x + {model.intercept_[0]:.2f}\nR^2 = {model.score(dataset[['YearsExperience']], dataset[['Salary']]):.3f}\nMSE = {mse:.2f}",
)
plt.show()
```
